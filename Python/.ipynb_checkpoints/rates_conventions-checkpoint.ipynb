# Справочник по типам начисления процентов и калькулятор конвенций

Этот ноутбук содержит математические формулы для различных типов начисления процентов (Compounding), их реализацию на «чистом» Python, а также аналогичные расчеты с использованием библиотеки **QuantLib** (`InterestRate.equivalentRate`).

---

## 1. Теоретический справочник и формулы

Пусть:
- $r$ — годовая процентная ставка (в долях единицы);
- $t$ — время в годах (Time to Maturity);
- $f$ — частота начисления процентов в год (Frequency: 1 — ежегодно, 2 — полугодовую, 4 — квартально и т.д.);
- $DF$ — дисконт-фактор (Discount Factor) — текущая стоимость 1 денежной единицы, выплачиваемой через время $t$.

### Типы начисления процентов и их формулы:

| Тип начисления | Код QuantLib | Формула Дисконт-Фактора ($DF$) | Базовое уравнение ставки ($r$) |
| :--- | :--- | :--- | :--- |
| **Simple** (Простое) | `ql.Simple` | $$DF = \frac{1}{1 + r \cdot t}$$ | $$r = \frac{1 - DF}{DF \cdot t}$$ |
| **Compounded** (Сложное дискретное) | `ql.Compounded` | $$DF = \frac{1}{\left(1 + \frac{r}{f}\right)^{f \cdot t}}$$ | $$r = f \cdot \left( DF^{-\frac{1}{f \cdot t}} - 1 \right)$$ |
| **Continuous** (Непрерывное) | `ql.Continuous` | $$DF = e^{-r \cdot t}$$ | $$r = -\frac{\ln(DF)}{t}$$ |
| **SimpleThenCompounded** | `ql.SimpleThenCompounded` | Если $t \le \frac{1}{f}$:<br>$$DF = \frac{1}{1 + r \cdot t}$$<br>Если $t > \frac{1}{f}$:<br>$$DF = \frac{1}{\left(1 + \frac{r}{f}\right)^{f \cdot t}}$$ | См. условия для $t$ выше. |

---

## 2. Логика перевода ставок

Перевод ставки из конвенции $A$ в конвенцию $B$ для заданного срока $t$ базируется на **принципе финансовой эквивалентности**: дисконт-факторы для обеих ставок должны быть абсолютно одинаковыми.

$$DF_A(r_A, t, f_A) = DF_B(r_B, t, f_B)$$

**Алгоритм конвертации на чистом Python:**
1. На основе исходной ставки $r_A$, её типа начисления и срока $t$ рассчитывается $DF$.
2. Из полученного $DF$ по аналитической формуле целевого типа начисления выражается новая ставка $r_B$.

```python
import math

def r_to_df(r, compounding, t, f=1):
    """Вычисляет Discount Factor на основе параметров ставки."""
    compounding = compounding.lower()
    if t <= 0:
        return 1.0
        
    if compounding == 'simple':
        return 1.0 / (1.0 + r * t)
        
    elif compounding == 'compounded':
        return 1.0 / math.pow(1.0 + r / f, f * t)
        
    elif compounding == 'continuous':
        return math.exp(-r * t)
        
    elif compounding == 'simplethencompounded':
        if t <= 1.0 / f:
            return 1.0 / (1.0 + r * t)
        else:
            return 1.0 / math.pow(1.0 + r / f, f * t)
    else:
        raise ValueError(f"Неизвестный тип начисления: {compounding}")

def df_to_r(df, compounding, t, f=1):
    """Выводит ставку r из заданного Discount Factor."""
    compounding = compounding.lower()
    if t <= 0:
        return 0.0
        
    if compounding == 'simple':
        # Уравнение: DF = 1 / (1 + r*t)
        return (1.0 - df) / (df * t)
        
    elif compounding == 'compounded':
        # Уравнение: DF = 1 / (1 + r/f)**(f*t)
        return f * (math.pow(df, -1.0 / (f * t)) - 1.0)
        
    elif compounding == 'continuous':
        # Уравнение: DF = exp(-r*t)
        return -math.log(df) / t
        
    elif compounding == 'simplethencompounded':
        if t <= 1.0 / f:
            return (1.0 - df) / (df * t)
        else:
            return f * (math.pow(df, -1.0 / (f * t)) - 1.0)
    else:
        raise ValueError(f"Неизвестный тип начисления: {compounding}")

def convert_rate_pure_python(rate, comp_from, f_from, comp_to, f_to, t):
    """Основная функция калькулятора (Чистый Python)"""
    # Шаг 1: Рассчитываем эквивалентный DF
    df = r_to_df(rate, comp_from, t, f_from)
    # Шаг 2: Из DF восстанавливаем целевую ставку
    target_rate = df_to_r(df, comp_to, t, f_to)
    return df, target_rate
```

## 3. Математическое описание решаемых уравнений

При конвертации, например, из **Compounded** ($r_1, f_1$) в **Continuous** ($r_2$), калькулятор неявно решает следующее уравнение относительно $r_2$:

$$e^{-r_2 \cdot t} = \frac{1}{\left(1 + \frac{r_1}{f_1}\right)^{f_1 \cdot t}} \implies -r_2 \cdot t = \ln\left( \left(1 + \frac{r_1}{f_1}\right)^{-f_1 \cdot t} \right) \implies r_2 = f_1 \cdot \ln\left(1 + \frac{r_1}{f_1}\right)$$

При конвертации из **Continuous** ($r_1$) в **Simple** ($r_2$):

$$\frac{1}{1 + r_2 \cdot t} = e^{-r_1 \cdot t} \implies 1 + r_2 \cdot t = e^{r_1 \cdot t} \implies r_2 = \frac{e^{r_1 \cdot t} - 1}{t}$$

## 4. Интерактивный калькулятор (Чистый Python)

Задайте входные параметры в ячейке ниже для проведения расчетов.

```python
# === ВХОДНЫЕ ДАННЫЕ ДЛЯ РАСЧЕТА ===
source_rate = 0.08         # 8% годовых
time_to_maturity = 0.75    # Срок: 9 месяцев (0.75 года)

# Допустимые типы: 'simple', 'compounded', 'continuous', 'simplethencompounded'
comp_source = 'compounded'
freq_source = 2            # Полугодовые выплаты (f=2)

comp_target = 'continuous'
freq_target = 1            # Для continuous частота формально не важна, передаем 1
# ==================================

df_res, rate_res = convert_rate_pure_python(
    source_rate, comp_source, freq_source, comp_target, freq_target, time_to_maturity
)

print(f"[Чистый Python] Результаты конвертации:")
print(f" Исходная ставка: {source_rate*100:.4f}% ({comp_source}, частота={freq_source})")
print(f" Рассчитанный Дисконт-Фактор (DF): {df_res:.6f}")
print(f" Конвертированная ставка: {rate_res*100:.4f}% ({comp_target}, частота={freq_target})")
```

## 5. Расчет с использованием QuantLib (`InterestRate.equivalentRate`)

Теперь повторим расчет с помощью встроенного механизма QuantLib. Метод `equivalentRate` объекта `InterestRate` принимает целевую конвенцию начисления, частоту и время, выполняя абсолютно идентичное сопоставление дисконт-факторов на низком уровне C++.

```python
import QuantLib as ql

# Словари для маппинга текстовых названий в энумы QuantLib
ql_comp_mapping = {
    'simple': ql.Simple,
    'compounded': ql.Compounded,
    'continuous': ql.Continuous,
    'simplethencompounded': ql.SimpleThenCompounded
}

ql_freq_mapping = {
    1: ql.Annual,
    2: ql.Semiannual,
    4: ql.Quarterly,
    12: ql.Monthly
}

def convert_rate_quantlib(rate, comp_from, f_from, comp_to, f_to, t):
    """Конвертация через QuantLib API"""
    # Выбираем энумы
    q_comp_from = ql_comp_mapping[comp_from.lower()]
    q_freq_from = ql_freq_mapping.get(f_from, ql.NoFrequency) if f_from > 1 else ql.Annual
    
    q_comp_to = ql_comp_mapping[comp_to.lower()]
    q_freq_to = ql_freq_mapping.get(f_to, ql.NoFrequency) if f_to > 1 else ql.Annual
    
    # Используем условный DayCounter (например, Actual365), так как нам важна лишь числовая доля года t
    dc = ql.Actual365Fixed()
    
    # Создаем объект исходной процентной ставки
    ir_source = ql.InterestRate(rate, dc, q_comp_from, q_freq_from)
    
    # Рассчитываем DF через QuantLib для проверки
    ql_df = ir_source.discountFactor(t)
    
    # Конвертируем в эквивалентную ставку
    ir_target = ir_source.equivalentRate(q_comp_to, q_freq_to, t)
    
    return ql_df, ir_target.rate()

# Запуск расчета через QuantLib по тем же параметрам
ql_df_res, ql_rate_res = convert_rate_quantlib(
    source_rate, comp_source, freq_source, comp_target, freq_target, time_to_maturity
)

print(f"[QuantLib] Результаты конвертации:")
print(f" Дисконт-Фактор (QuantLib DF): {ql_df_res:.6f}")
print(f" Эквивалентная ставка (QuantLib): {ql_rate_res*100:.4f}%")
print(f"\nСверка результатов (Разница): {abs(rate_res - ql_rate_res):.2e}")
```
